```python
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
```

```python
pluck1, sr = librosa.load('/Users/adamdeedman/Documents/University/Lectures/Third Year/FYP/Python Stage/IDMT-SMT-BASS/PS/FS/BS_1_EQ_1_FS_NO_1_10.wav')

# load sound file (and sample rate)
```

```python
ipd.Audio(data=pluck1, rate=sr)

# play sound file
```

```python
filter_banks = librosa.filters.mel(n_fft=2048, sr=sr, n_mels=10)
filter_banks.shape

# mel filter banks
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(filter_banks, sr=sr, x_axis="linear")
plt.colorbar(format="%+2.f")
plt.show()

# plot mel bands to visualise
```

```python
mel_spectogram = librosa.feature.melspectrogram(pluck1, n_fft=2048, sr=sr, hop_length=512, n_mels = 10) # extracts mel spectogram
log_mel_spectogram = librosa.power_to_db(mel_spectogram) # converts mel spectrogram to logarithm, using dB
plt.figure(figsize=(25,10))
librosa.display.specshow(log_mel_spectogram, x_axis="time", y_axis="mel", sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```
