```python
import essentia.standard as es
import numpy as np
import matplotlib.pyplot as plt
import librosa
```

```python
# Load audio files
pluck1 = es.MonoLoader(filename='/Users/adamdeedman/Documents/University/Lectures/Third Year/FYP/Python Stage/IDMT-SMT-BASS/PS/FS/BS_1_EQ_1_FS_NO_1_10.wav')()
pick1 = es.MonoLoader(filename='/Users/adamdeedman/Documents/University/Lectures/Third Year/FYP/Python Stage/IDMT-SMT-BASS/PS/PK/BS_1_EQ_1_PK_NO_1_10.wav')()
slap1 = es.MonoLoader(filename='/Users/adamdeedman/Documents/University/Lectures/Third Year/FYP/Python Stage/IDMT-SMT-BASS/PS/ST/BS_1_EQ_1_ST_NO_1_10.wav')()
```

```python
# Make sure audio signals have even number of samples
if len(pluck1) % 2 == 1:
    pick1 = np.append(pluck1, 0)


if len(pick1) % 2 == 1:
    pick1 = np.append(pick1, 0)


if len(slap1) % 2 == 1:
    pick1 = np.append(slap1, 0)
```

```python
# Set frame size and hop size

frame_size = 2048
hop_size = 1024
```

```python
# Calculate magnitude spectrum frames

pluck1_spec = es.Spectrum(size=frame_size)  # Initialize spectrum algorithm
pluck1_frame_generator = es.FrameGenerator(pluck1, frameSize=frame_size, hopSize=hop_size, startFromZero=True)  # Initialize frame generator
pluck1_spec_frames = [pluck1_spec(frame) for frame in pluck1_frame_generator]

pick1_spec = es.Spectrum(size=frame_size)  # Initialize spectrum algorithm
pick1_frame_generator = es.FrameGenerator(pick1, frameSize=frame_size, hopSize=hop_size, startFromZero=True)  # Initialize frame generator
pick1_spec_frames = [pick1_spec(frame) for frame in pick1_frame_generator]

slap1_spec = es.Spectrum(size=frame_size)  # Initialize spectrum algorithm
slap1_frame_generator = es.FrameGenerator(slap1, frameSize=frame_size, hopSize=hop_size, startFromZero=True)  # Initialize frame generator
slap1_spec_frames = [slap1_spec(frame) for frame in slap1_frame_generator]
```

```python
# Calculate central moments, spread, skewness, and kurtosis for each frame

pluck1_features = []
for frame in pluck1_spec_frames:
    pluck1_spread, pluck1_skewness, pluck1_kurtosis = es.DistributionShape()(es.CentralMoments()(frame))
    pluck1_features.append([pluck1_spread, pluck1_skewness, pluck1_kurtosis])

pick1_features = []
for frame in pick1_spec_frames:
    pick1_spread, pick1_skewness, pick1_kurtosis = es.DistributionShape()(es.CentralMoments()(frame))
    pluck1_features.append([pick1_spread, pick1_skewness, pick1_kurtosis])

slap1_features = []
for frame in slap1_spec_frames:
    slap1_spread, slap1_skewness, slap1_kurtosis = es.DistributionShape()(es.CentralMoments()(frame))
    slap1_features.append([slap1_spread, slap1_skewness, slap1_kurtosis])
```
