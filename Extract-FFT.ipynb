```python
# Get fourier co-efficients for each signal

pluck1_ft = np.fft.fft(pluck1)
pick1_ft = np.fft.fft(pick1)
slap1_ft = np.fft.fft(slap1)
```

```python
# To get just the magnitude value, the absolute value of the fourier co-efficient is taken

magspec_pluck1 = np.abs(pluck1_ft)
magspec_pick1 = np.abs(pick1_ft)
magspec_slap1 = np.abs(slap1_ft)
```

```python
# Function to extract and plot FFT of signal

def plot_magspec(signal, title, sr, f_ratio=1):
    ft = np.fft.fft(signal)
    magspec = np.abs(ft)

    # plot magnitude spectrum
    plt.figure(figsize=(13,5))
    
    freqaxis = np.linspace(0, sr, len(magspec))
    numfreqbins = int(len(freqaxis) * f_ratio) # f_ratio allows you to stop the fft plotting the entire range of frequencies

    plt.plot(freqaxis[:numfreqbins], magspec[:numfreqbins])
    plt.xlabel("Frequency (Hz)")
    plt.title(title)

    plt.show()
```

```python
plot_magspec(pluck1, "pluck1 FFT", sr, 0.1)
```

```python
plot_magspec(pick1, "pick1 FFT", sr, 0.1)
```

```python
plot_magspec(slap1, "slap1 FFT", sr, 0.1)
```

Interestingly, the pluck has the least powerful harmonics, the pick has more powerful harmonics, but in the slap signal, the second harmonic is stronger than the fundamental.
